```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_gbm(S0, mu, sigma, T, dt):
    """
    Simulate a Geometric Brownian Motion (GBM) path.

    Parameters:
    - S0: initial asset price
    - mu: drift coefficient (expected return)
    - sigma: volatility coefficient
    - T: time horizon
    - dt: time step size

    Returns:
    - t: array of time points
    - S: array of simulated asset prices
    """
    N = int(T / dt)  # number of time steps
    t = np.linspace(0, T, N)
    S = np.zeros(N)
    S[0] = S0

    for i in range(1, N):
        Z = np.random.standard_normal()  # random draw from standard normal distribution
        S[i] = S[i-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

    return t, S
```

```python
def test_gbm():
    S0 = 100  # initial asset price
    mu = 0.05  # expected return
    sigma = 0.2  # volatility
    T = 1  # 1 year
    dt = 1/252  # daily time step

    t, S = simulate_gbm(S0, mu, sigma, T, dt)

    # Plot the results
    plt.plot(t, S)
    plt.title("Geometric Brownian Motion Simulation")
    plt.xlabel("Time (years)")
    plt.ylabel("Asset Price")
    plt.show()

# Run the test
test_gbm()
```
